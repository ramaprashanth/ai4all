## What is Jupyter Notebook? 
Jupyter notebook allows you to interactively create text and code blocks that are viewable from a web browser and exportable to pdf, html and other popular formats.

In addition to python, you can also use Jupyter Notebook for R, Spark and Julia.

It's a great tool for:

- Sharing the steps taken in your code
- Creating documentation of data analysis
- Outputting visualizations and reports
- It's not the right tool for creating programs.

## Starting Jupyter Notebook
Jupyter Notebook runs inside your browser window. 

### Mac:
- Open your terminal (space bar + command, then search for terminal)
- On the command line, type `jupyter notebook`
- Jupyter Notebook will open in your browser window

### Windows
- Open Powershell or your command prompt (windows button + r, then type cmd)
- Type `jupyter notebook`
- Jupyter Notebook will open from in a browser window

On windows, if you installed python via Anaconda, you may also have some program options under Anaconda to open Jupyter Notebook that way

You will be taken to a directory tree where you can:

- view files in current directory
- create new notebooks
- edit or delete notebooks
- view running notebooks

By default, your ipython notebooks can be found by going to the following URL in your browser: http://127.0.0.1:8888/

If you get a note about logging in, then you need to look at the terminal window that is running jupyter notebook and copy and paste the token from that window into the URL.

## Create a new notebook
From the main notebook page:

- Click on "New"
- Choose Python 3
When Jupyter loads, it will have a menu bar on the top and then an editor bar.   After that, it will have the first 'Cell' or 'Block' in the notebook that allows you to enter code, markup or text.

<img src="jupyter_start.png">

The 'In [ ]:' prompt says that the block is waiting for input.  The area between the square brackets is where it will display status indicators on what state the block is in - waiting, processing, or finished.

1. Create a text (markdown) block
To create text blocks such as this one, from the "code" dropdown, select "markdown".

Copy the following text into your ipython notebook:

```python
2 + 2
```

```python
# Header 1  

## Header 2  

### Header 3  

** bold stuff **  

*italic stuff*  

[link name](http://www.imadethisup.com)  

* bullet 1  

    * bullet 2  
    
        * bullet 3
        
1. numbered list 1  
2. numbered list 2  
3. numbered list 3  
```

2. Run the markdown block

Click somehwere inside the markdown block that you copied and pasted from the text above and hit "Shift + Enter" on your keyboard.

Notice that the formatting changes, letting you know that the block has finished running.

3. Create a code block
With Jupyter notebook, you can run code right on your page. Copy and paste the following code in a code block:

`print("Hello world! Now we're getting fancy!")`

To run the code block above, click somewhere in the block and then shit "Shift + Enter" on your keyboard.  

Notice that the print statement appears, and that a number appears in the square brackets next to the 'In [ ]:'.  The number indicates that the block has finished running and what order it was run in compared with other blocks on the page.
